```
# 4. Esponi con LocalTunnel (NESSUNA REGISTRAZIONE!)
import subprocess, re, time

# Avvia localtunnel in background
proc = subprocess.Popen(
    ['npx', 'localtunnel', '--port', '11434'],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True
)

# Leggi l'URL
time.sleep(5)
for line in proc.stdout:
    if 'your url is' in line.lower():
        url = line.strip().split()[-1]
        break

print("="*60)
print("🚀 COPIA QUESTO URL NELL'EDITOR ELAB:")
print(f"\n   {url}\n")
print("="*60)
print("\nIn useFunctionGemma.js linea 415:")
print(f"   baseUrl: '{url}'")
```

```
# 5. Test velocità GPU
import requests, time
start = time.time()
try:
    r = requests.post(f"{url}/api/generate", json={
        "model": "gemma2:2b",
        "prompt": 'Rispondi JSON: {"test": "ok"}',
        "stream": False,
        "options": {"num_predict": 50}
    }, timeout=30)
    print(f"⚡ Risposta in {time.time()-start:.1f}s")
    print(f"Output: {r.json().get('response', 'ERROR')[:100]}")
except Exception as e:
    print(f"❌ Errore: {e}")
    print("Prova a cliccare il link URL sopra e poi clicca 'Click to Continue'")
```
